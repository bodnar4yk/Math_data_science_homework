### 6.5 Создайте гистограмму цен для самого распространенного типа комнат в Амстердаме. С какими проблемами вы столкнулись?

```python
select=data[ (data['room_type']=='Entire home/apt')]
decr=select[["price"]].describe()
decr
#75% данных иммеет цену до 200. макс 8000.
#75% значений будут в левом углу имало читаемые, как видно на графике ниже.
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 5))
sns.despine(fig)

sns.histplot(
    data,
    x="price",
    multiple="stack",
    palette="light:m_r" )

ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
ax.set_xticks([100,150,200,300, 400, 500, 1000, 2000, 5000, 8000])
```

```python
select = data[ (data['room_type']=='Entire home/apt') & (data['price']<200)] #данные c ценой до 200, 75% от всех данных
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
sns.despine(fig)

sns.histplot(
    select,
    x="price",
    multiple="stack",
    palette="light:m_r" )

ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
ax.set_xticks([50,100,150,200])
#наиболее популярные комнаты в ценовом диапазоне приблизительно от 100 до 150
```

```python
#так же можно посмотреть на tail (price>200)
select_2 = data[ (data['room_type']=='Entire home/apt') & (data['price']>200)] #данные c ценной более 200, 25% от всех данных
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
sns.despine(fig)

sns.histplot(
    select_2,
    x="price",
    multiple="stack",
    palette="light:m_r" )

ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
ax.set_xticks([200,300,400,500,1000,2000, 8000]) 
#трудно читаемый график
```
