## Стандартное распределение 

### Столы и стулья

Ваша фирма получила заказ на столы и стулья для старшей школы. Любое производство подвержено неточностям. Вы решили моделировать высоту стульев, как случайную величину со стандартным распределением, с ожиданием 83 см и среднеквадратическим отклонением 5 см. 

1. Какая вероятность, что высота произведенного стула не меньше 82 см?

2. Какая минимальная высота стульев, которые составляют 20% самых высоких стульев?

3. Какая нижняя и высшая граница 95% центрального интервала? Что это означает?

```python
mu = 83; sigma = 5
mean, variance, kurtosis  = stats.norm.stats(mu,sigma, moments='mvk')
```

```python
stats.norm.median(mu,sigma)
x = np.linspace(stats.norm.ppf(0.01, mu,sigma),stats.norm.ppf(0.99, mu,sigma),100) # linspace instead of range!
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, stats.norm.pdf(x, mu,sigma), 'b-', ms=2, label='gauss pdf')
```

```python
#Какая вероятность, что высота произведенного стула не меньше 82 см?
1-stats.norm.cdf(82, mu,sigma)

stats.norm.sf(82, mu,sigma) #інший варіант
```

```python
#Какая минимальная высота стульев, которые составляют 20% самых высоких стульев?
stats.norm.ppf(0.8, mu,sigma)
```

```python
#Какая нижняя и высшая граница 95% центрального интервала?
stats.norm.interval(0.95, mu,sigma)
```

## Распределение Пуассона

Завод считает количество поломок станка в месяц с помощью распределения Пуассона с параметером Лямбда=2

1. С какой вероятностью станок не будет иметь ни одной поломки в месяц?

2. С какой вероятностью станок будет иметь больше 2 поломок в месяц?

3. Сколько поломок за 3 месяца ожидаются с вероятностью 90%?

4. Видите ли вы проблемы в моделировании этой ситуации распределением Пуассона?

```python
lamb = 2
mean, variance, kurtosis  = stats.poisson.stats(lamb, moments='mvk')
x = np.arange(stats.poisson.ppf(0.01, lamb),stats.poisson.ppf(0.99, lamb)) ## x как квантили от вероятностей
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, stats.poisson.pmf(x, lamb), 'bo', ms=8, label='poisson pmf')
ax.vlines(x, 0, stats.poisson.pmf(x, lamb), colors='b', lw=5, alpha=0.5);
```

```python
#С какой вероятностью станок не будет иметь ни одной поломки в месяц?
stats.poisson.pmf(0,lamb)
```

```python
#С какой вероятностью станок будет иметь больше 2 поломок в месяц?
1-stats.poisson.pmf(0,lamb)-stats.poisson.pmf(1,lamb)-stats.poisson.pmf(2,lamb)

stats.poisson.sf(2,lamb) #інший варіант
```

```python
#Сколько поломок за 3 месяца ожидаются с вероятностью 90%?
lamb_3=3*lamb
stats.poisson.ppf(0.9, lamb_3)
```

```python
#Видите ли вы проблемы в моделировании этой ситуации распределением Пуассона?
впродовж 3 місяців може статись подія, яка може повпливати на наше обчислення (через значення p або n, що змінить ламбду), 
але зараз у розподілі ми це не побачимо. 

Тобто ми припускаємо, що впродовж 3 місяців нічого не зміниться, тоді ми довіряємо даному результату. 
```
